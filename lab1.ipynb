```python
with open("dna_raw_complicated.txt", "r") as dna_text:
    ls = dna_text.readlines()
    
    print(ls)

    # Replaces '\n' with nothing if its not part of a '>seq' line.
    dna_strs = [i.replace("\n","") if ">" not in i else i for i in ls]

    # Joining it as on single string
    string = "".join(dna_strs)

    #Splits the string based on where '\n' is. INto a list.
    dna_strs = string.split("\n")

    print(dna_strs)

    # Removes the first 5 letters.
    dna_strs = [i[:-5] if ">" in i else i for i in dna_strs][1:]

    print(dna_strs)
    

```

```python
dna_dct_lst = []
for dna_str in dna_strs:
    #A, T, C and G
    a = dna_str.lower().count("a")
    t = dna_str.lower().count("t")
    c = dna_str.lower().count("c")
    g = dna_str.lower().count("g")
    
    dct = {"a":a, "t": t, "c": c, "g":g}

    dna_dct_lst.append(dct)

print(dna_dct_lst)

```

```python
import matplotlib.pyplot as plt

for i in range(len(dna_dct_lst)):
    lables = list(dna_dct_lst[i].keys())
    values = list(dna_dct_lst[i].values())


    plt.bar(lables, values, color="blue")

    # Lgg till titlar och etiketter
    plt.title("DNA Sequence Frequency")
    plt.xlabel("DNA Letters")
    plt.ylabel("Frequency")
    plt.show()

```
